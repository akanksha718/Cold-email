```python
import chromadb
client = chromadb.Client()
collection = client.get_or_create_collection(name="my_collection")
```

```python
collection.add(
    documents=["This is a about New York.", "This is about San Francisco."],
    ids=["doc1", "doc2"]
)
```

```python
all_docs = collection.get()
print(all_docs)
```

```python
results = collection.query(
    query_texts=["Tell me about silicon valley."],
    n_results=2
)
print(results)
```
